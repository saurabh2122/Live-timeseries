**EDA with Timeseries Data**



1.   Importing Pandas and Yfinance
2.   Yfinance is also known as Yahoo finance which is Python Library for retrieving historical stock data

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

we are working on Amazon Share data
where 'AMZN' is the ticker refers to a unique series of letters of Amazon share.

```python
amazon = yf.Ticker("AMZN")
```

```python
amazon_history = amazon.history(period="max")
```

Converting Into Data Frame

```python
amazon_df = pd.DataFrame(amazon_history)
```

Getting overview of dataframe by head function which returns to first few records

```python
amazon_df.head()
```

Getting overview of dataframe by tail function which returns to latest few records

```python
amazon_df.tail()
```

Ploting a simple line graph for high column

```python
amazon_df['High'].plot(figsize=(12,4)),
```

focusing on particular trend from year 2020 till 2022 by defining xlim which focuses on x axis

```python
amazon_df['High'].plot(xlim=['2020-01-01','2022-01-01'],figsize=(12,4))
```

Providing Ylim for even more better understanding

```python
amazon_df['High'].plot(xlim=['2020-01-01','2022-01-01'],ylim=[80,200],figsize=(12,4))
```



1.   Here ls is used to specifiy the type of line graph you can have. ( **'--'** , **'.-'** , **':'** )are some of types.
2.   c specify the color you can give to line



```python
amazon_df['High'].plot(xlim=['2020-01-01','2022-01-01'],ylim=[80,200],figsize=(12,4),ls=':',c='red')
```

**Time Sampling**

```python
amazon_df.index = pd.to_datetime(amazon_df['Date'])
```

```python
amazon_df
```

We are Using Aggregate Function Min for retrieving all time low stock value by end of the year

```python
amazon_df['Open'].resample(rule='A').min().plot()
```

We are Using Aggregate Function Max for retrieving all time low stock value by end of the year

```python
amazon_df.resample(rule='A').max()['Open'].plot()
```

```python
amazon_df['Open'].resample(rule='Y').max().plot(kind='bar').figsize=18,10
```

```python
amazon_df['Open'].rolling(10).mean().head(20)
```

Adding A column (Open: 30 Days Rolling) for storing Rolling mean values of 'Open' Column

```python
amazon_df['Open: 30days Rolling ']= amazon_df['Open'].rolling(10).mean()
```

```python
amazon_df.tail(20)
```

```python

amazon_df[['Open','Open: 30days Rolling ']].plot(xlim=['2020-03-19','2024-03-19'],ylim=[75,200],figsize=(12,4))
```
